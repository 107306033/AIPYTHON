# 下禮拜台北市氣溫預報圖（Apparent Temperature，）
### （資料由網路上查詢）

```python
%pylab inline
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import numpy as np
```

```python
date = ['3/16','3/17','3/18','3/19','3/20','3/21','3/22']
highApparentTemp = [20,24,25,25,25,28,29]
lowApparentTemp = [11,15,18,18,18,19,19]
plt.figure(figsize=(5,2.5),dpi=100,linewidth = 0.5)
plt.plot(date,highApparentTemp,'s-',color = 'r', label="High Apparent Temperature")
plt.plot(date,lowApparentTemp,'s-',color = 'g', label="Low Apparent Temperature")
plt.title("Apparent Temperature in Taipei next week", x=0.5, y=1.03)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlabel("Date", fontsize=7.5, labelpad = 7.5)
plt.ylabel("Apparent Temperature(Celsius)", fontsize=7.5, labelpad = 10)
plt.legend(loc = "best", fontsize=7.5)
plt.show()
```

```python
uv = [6,5,4,4,5,5,7]
date = ['3/16','3/17','3/18','3/19','3/20','3/21','3/22']
plt.bar(range(7),uv, align = 'center',color= 'steelblue', alpha = 0.8)
plt.ylabel('UV Amount') 
plt.title('UV Amount in Taipei next week')
plt.xticks(range(7),date)
plt.ylim([0,9])
plt.show()
```
